## The if statement

In order to write useful programs, we almost always need the ability to check conditions and change the behavior of the program accordingly. Conditional statements give us this ability. The simplest form is the if statement, which has the genaral form:

```python
if BOOLEAN EXPRESSION:
    STATEMENTS
```

A few important things to note about if statements:

1. The colon (:) is significant and required. It separates the header of the compound statement from the body.
2. The line after the colon must be indented. It is standard in Python to use four spaces for indenting.
3. All lines indented the same amount after the colon will be executed whenever the BOOLEAN_EXPRESSION is true.

Here is an example:

```python
food = 'spam'

if food == 'spam':
    print('Ummmm, my favorite!')
    print('I feel like saying it 100 times...')
    print(100 * (food + '! '))
```

The boolean expression after the if statement is called the condition. If it is true, then all the indented statements get executed. What happens if the condition is false, and food is not equal to 'spam'? In a simple if statement like this, nothing happens, and the program continues on to the next statement.

## Conditionals:

Decision making is very critical to any system. We want to be able to code logic. Instruct the computer on what to do given certain conditions or otherwise.

The syntax for decisions in python is:

### if expression: statement else: another statement

The flow chart is given below:

source: http://cess.nyu.edu/cess-experiments/z-tree-cheat-sheet/ii-conditional-statements-and-loops/

```python
if BOOLEAN EXPRESSION:
    STATEMENTS_1        # executed if condition evaluates to True
else:
    STATEMENTS_2        # executed if condition evaluates to False
```

```python
if True:          # This is always true
    pass          # so this is always executed, but it does nothing
else:
    pass
```

```python
# example the if else statement in python
morning = "good"
if morning == "good":
    print ("Good Morning")  # executes if condition is true
else:
    print ("You had a nightmare?")  # executes if condition is false
```

## Chained Conditionals
There are situations where we need to capture more than one condition. The flow chart looks like this:

```python
if x < y:
    STATEMENTS_A
elif x > y:
    STATEMENTS_B
else:
    STATEMENTS_C
```

```python
# example
item = "infinite"
if item == "flat":
    print ("it is a book") # does this if item is flat
elif item == "spherical":
    print ("It is the earth") # does this if item is spherical
elif item == "infinite":
    print ("It is the universe") # does this if item is infinite
else: 
    print ("I don't know what it is") # does this if any of the above is not true
```

## Nested Conditionals
There are also situations where you need a another condition inside a condition.

```python
if x < y:
    STATEMENTS_A
else:
    if x > y:
        STATEMENTS_B
    else:
        STATEMENTS_C
```

```python
# example
age =  20
if age > 40:
    print ("I am Grown up") 
else:
    if age < 13:    # nested if block
        print ("I am a child")
    else:
        print ("I am a youth")
```
